```python
# Import Library yang akan digunakan
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

**LANGKAH 3 MENYIAPKAN DATASET**

```python
# Menyiapkan data dan memanggil dataset
dataset = pd.read_csv('drive/My Drive/Colab Notebooks/konsumen.csv')
dataset.keys()
```

```python
# Menampilkan 5 baris data pertama dari dataset tersebut
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# Konversi ke data array
X = np.asarray(dataset)
print(X)
```

```python
# Menampilkan data dalam bentuk scatter plot
plt.scatter(X[:, 0], X[:, 1], label='True Position')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Konsumen")
plt.legend()
plt.show()
```

**LANGKAH 4 MENGGUNAKAN LIBRARY K-MEANS**

```python
# Mengaktifkan K-Means dengan jumlah K = 2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
# Menampilkan nilai centroid yang di-generate oleh algoritma
print(kmeans.cluster_centers_)
```

```python
# Menampilkan label data point
print(kmeans.labels_)
```

**LANGKAH 5 MENAMPILKAN OUTPUT**

```python
# Plot data point
# Memvisualisasikan bagaimana telah di-klasterisasi
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Konsumen")
plt.show()
```

```python
# Plot data point
# Memvisualisasikan bagaimana telah di-klasterisasi.
# Menampilkan centroids dengan warna hitam.
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Konsumen")
plt.show()
```
