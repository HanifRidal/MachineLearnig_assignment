```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


import simpsom as sps
```

```python
url = 'https://raw.githubusercontent.com/kokocamp/vlog119/refs/heads/main/vlog119.csv'
vlog139 = pd.read_csv(url)
```

```python
X = vlog139[['gpa', 'gmat', 'work_experience', 'admitted']]
y = vlog139[['admitted']]

scaler = StandardScaler()
data = scaler.fit_transform(pd.DataFrame(X))
labels = scaler.fit_transform(pd.DataFrame(y))

print(data)
```

```python
net = sps.SOMNet(10, 10, data)#inisialisasi lebar dan panjang neuron, serta data nya
net.train(start_learning_rate=0.01,epochs=1000) #untuk training dengan parameternya
net.nodes_graph()
net.diff_graph()
net.project(data, labels=labels)
net.cluster(data)
```
